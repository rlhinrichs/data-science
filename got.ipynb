```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Use Pandas to load the `got.csv' dataset.
data = pd.read_csv("got.csv")
pd.DataFrame(data)
```

```python
# Create a series containing the number of times each king was an attacker.
attackers = data["attacker_king"]
attackers.value_counts()
```

```python
# Create a series containing the number of times each king was a defender.
defenders = data["defender_king"]
defenders.value_counts()
```

```python
# Combine these two variables into a single series. 
# (Hint: How should you combine these two series?
# Can you add series in Pandas?)
kings = {'attackers': attackers.value_counts(),'defenders': defenders.value_counts()}
battles = pd.DataFrame(kings).fillna(0)
battles
```

```python
# Use your combined data to retrieve labels for your x-axis ticks.
# Create a bar chart with red bars and set its xô€€€axis tick labels appropriately.
# Add a title and labels to your plot.
battles.plot(kind="bar",ylabel="Numbers of Battles",title="Kings who Fought")
```
